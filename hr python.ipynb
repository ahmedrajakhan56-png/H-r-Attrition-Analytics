```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")


print("📊 HR ATTRITION ANALYSIS - IBM DATASET")

# CORRECT PATH
file_path = r"C:\Users\ACER\Desktop\HR Data.csv"
df = pd.read_csv(file_path)

print(f"\n✅ Dataset loaded!")
print(f"📈 Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\n📋 Columns:\n{df.columns.tolist()}")
```

```python


print("📊 INITIAL DATA EXPLORATION")


print("\n📋 Dataset Info:")
print(df.info())

print("\n🔍 Missing Values:")
print(df.isnull().sum()[df.isnull().sum() > 0] if any(df.isnull().sum() > 0) else "✅ No missing values!")

print("\n🎯 Attrition Distribution:")
att_counts = df['Attrition'].value_counts()
att_pct = df['Attrition'].value_counts(normalize=True) * 100
print(f"Yes: {att_counts['Yes']} ({att_pct['Yes']:.1f}%)")
print(f"No: {att_counts['No']} ({att_pct['No']:.1f}%)")
```

```python



print("🧹  CLEANING")


df_clean = df.copy()

# Attrition to binary
df_clean['Attrition_Flag'] = df_clean['Attrition'].map({'Yes': 1, 'No': 0})
print("✅ Attrition_Flag created")


# Drop redundant columns
df_clean.drop(columns=['CF_attrition label', 'Over18', 'Employee Count', 'Standard Hours', '-2', '0'], 
              inplace=True, errors='ignore')
print("✅ Redundant columns dropped")

# Employee ID
df_clean['EmployeeID'] = df_clean['Employee Number']
print("✅ EmployeeID created")

# Job Role clean
df_clean['JobRole_Clean'] = df_clean['Job Role'].str.strip()
print("✅ JobRole_Clean created")

# Marital Status clean
df_clean['MaritalStatus_Clean'] = df_clean['Marital Status'].str.strip()
print("✅ MaritalStatus_Clean created")

# Overtime flag
df_clean['OverTime_Flag'] = df_clean['Over Time'].map({'Yes': 1, 'No': 0})
print("✅ OverTime_Flag created")

# Employment Status
status_map = {1: 'Current', 0: 'Ex-Employee'}
df_clean['EmploymentStatus'] = df_clean['CF_current Employee'].map(status_map)
print("✅ EmploymentStatus created")

# Education Level
edu_map = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
df_clean['EducationLevel'] = df_clean['Education'].map(edu_map)
print("✅ EducationLevel created")

# Satisfaction levels
sat_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df_clean['EnvironmentSat_Level'] = df_clean['Environment Satisfaction'].map(sat_map)
df_clean['JobSat_Level'] = df_clean['Job Satisfaction'].map(sat_map)
df_clean['RelationshipSat_Level'] = df_clean['Relationship Satisfaction'].map(sat_map)
print("✅ Satisfaction levels created")

# Job Involvement
inv_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df_clean['JobInvolvement_Level'] = df_clean['Job Involvement'].map(inv_map)
print("✅ JobInvolvement_Level created")

# Job Level category
level_map = {1: 'Entry', 2: 'Junior', 3: 'Mid', 4: 'Senior', 5: 'Executive'}
df_clean['JobLevel_Category'] = df_clean['Job Level'].map(level_map)
print("✅ JobLevel_Category created")

# Salary in Lakhs
df_clean['MonthlyIncome_Lakhs'] = df_clean['Monthly Income'] / 100000
print("✅ MonthlyIncome_Lakhs created")

# Salary bands
bins_salary = [0, 20000, 30000, 40000, 50000, 70000, 100000, 200000]
labels_salary = ['<20K', '20K-30K', '30K-40K', '40K-50K', '50K-70K', '70K-1L', '>1L']
df_clean['SalaryBand'] = pd.cut(df_clean['Monthly Income'], bins=bins_salary, labels=labels_salary, right=False)
print("✅ SalaryBand created")

# Companies worked groups
bins_companies = [0, 1, 2, 3, 5, 10]
labels_companies = ['1', '2', '3', '4-5', '6+']
df_clean['CompaniesWorked_Group'] = pd.cut(df_clean['Num Companies Worked'], bins=bins_companies, labels=labels_companies, right=False)
print("✅ CompaniesWorked_Group created")

# Performance Rating
perf_map = {3: 'Fully Meets', 4: 'Exceeds'}
df_clean['PerformanceRating_Clean'] = df_clean['Performance Rating'].map(perf_map)
df_clean['HighPerformer'] = (df_clean['Performance Rating'] >= 4).astype(int)
print("✅ PerformanceRating_Clean & HighPerformer created")

# Stock Options
df_clean['HasStockOptions'] = (df_clean['Stock Option Level'] > 0).astype(int)
print("✅ HasStockOptions created")

# Work Life Balance
wlb_map = {1: 'Bad', 2: 'Average', 3: 'Good', 4: 'Excellent'}
df_clean['WorkLifeBalance_Level'] = df_clean['Work Life Balance'].map(wlb_map)
print("✅ WorkLifeBalance_Level created")

# Tenure groups
bins_tenure = [0, 1, 3, 5, 10, 15, 20, 40]
labels_tenure = ['<1 year', '1-3 years', '3-5 years', '5-10 years', '10-15 years', '15-20 years', '20+ years']
df_clean['TenureGroup'] = pd.cut(df_clean['Years At Company'], bins=bins_tenure, labels=labels_tenure, right=False)
print("✅ TenureGroup created")

print(f"\n✅ Cleaning complete! Shape: {df_clean.shape}")
```

```python



print("🔧 FEATURE ENGINEERING")


# Age groups
bins_age = [18, 25, 35, 45, 55, 65]
labels_age = ['18-25', '26-35', '36-45', '46-55', '56-65']
df_clean['AgeGroup'] = pd.cut(df_clean['Age'], bins=bins_age, labels=labels_age, right=False)
print("✅ AgeGroup created")

# Tenure flags
df_clean['LongTenure'] = (df_clean['Years At Company'] > 5).astype(int)
df_clean['NewHire'] = (df_clean['Years At Company'] < 1).astype(int)
print("✅ LongTenure & NewHire created")

# Average Satisfaction
sat_cols = ['Environment Satisfaction', 'Job Satisfaction', 'Relationship Satisfaction']
df_clean['AvgSatisfaction'] = df_clean[sat_cols].mean(axis=1)
df_clean['HighSatisfaction'] = (df_clean['AvgSatisfaction'] > 3).astype(int)
df_clean['LowSatisfaction'] = (df_clean['AvgSatisfaction'] <= 2).astype(int)
print("✅ Satisfaction features created")

# Company flags
df_clean['MultipleCompanies'] = (df_clean['Num Companies Worked'] > 3).astype(int)
df_clean['FirstJob'] = (df_clean['Num Companies Worked'] <= 1).astype(int)
print("✅ Company flags created")

# Involvement flags
df_clean['HighInvolvement'] = (df_clean['Job Involvement'] >= 3).astype(int)
df_clean['LowInvolvement'] = (df_clean['Job Involvement'] <= 2).astype(int)
print("✅ Involvement flags created")

# Work-life balance flags
df_clean['GoodWorkLife'] = (df_clean['Work Life Balance'] >= 3).astype(int)
df_clean['PoorWorkLife'] = (df_clean['Work Life Balance'] <= 2).astype(int)
print("✅ Work-life balance flags created")

# Salary flags
df_clean['HighPaid'] = (df_clean['Monthly Income'] > 50000).astype(int)
df_clean['LowPaid'] = (df_clean['Monthly Income'] < 30000).astype(int)
print("✅ Salary flags created")

print(f"\n✅ Total new features: 15")
```

```python



print("💡 KEY INSIGHTS")


total_emp = len(df_clean)
att_count = df_clean['Attrition_Flag'].sum()
att_rate = (att_count/total_emp)*100
replacement_cost = 500000
total_loss = att_count * replacement_cost

print(f"""
📊 CURRENT STATE:
├── Total Employees: {total_emp}
├── Attrition: {att_count} ({att_rate:.1f}%)
├── Replacement Cost: ₹5 Lakhs
└── Annual Loss: ₹{total_loss:,.0f} (₹{total_loss/10000000:.1f} Cr)

🔍 KEY DRIVERS:
""")

# Overtime
ot_risk = df_clean.groupby('Over Time')['Attrition_Flag'].mean()*100
print(f"\n⏰ OVERTIME: {ot_risk['Yes']:.1f}% vs {ot_risk['No']:.1f}% ({ot_risk['Yes']/ot_risk['No']:.1f}x)")

# Salary
low_sal = df_clean[df_clean['SalaryBand'] == '<20K']['Attrition_Flag'].mean()*100 if '<20K' in df_clean['SalaryBand'].values else 0
high_sal = df_clean[df_clean['SalaryBand'] == '>1L']['Attrition_Flag'].mean()*100 if '>1L' in df_clean['SalaryBand'].values else 0
print(f"💰 SALARY: Low: {low_sal:.1f}% | High: {high_sal:.1f}%")

# Department
dept_risk = df_clean.groupby('Department')['Attrition_Flag'].mean()*100
print(f"\n🏢 DEPARTMENT:")
for dept, rate in dept_risk.sort_values(ascending=False).items():
    print(f"  {dept}: {rate:.1f}%")

# Performance
perf_risk = df_clean.groupby('PerformanceRating_Clean')['Attrition_Flag'].mean()*100
print(f"\n⭐ PERFORMANCE:")
for perf, rate in perf_risk.items():
    if pd.notna(perf):
        print(f"  {perf}: {rate:.1f}%")

# Potential savings
savings = int(att_count * 0.15 * replacement_cost)
print(f"\n💰 POTENTIAL SAVINGS: ₹{savings:,.0f} (15% reduction)")
```

```python
# ============================================
# CELL 6: SAVE FOR POWER BI
# ============================================


print("💾 SAVING FOR POWER BI")


# Essential columns
essential = [
    'Age', 'AgeGroup', 'Attrition', 'Attrition_Flag', 'BusinessTravel_Clean',
    'Department', 'Education', 'EducationLevel', 'JobRole_Clean',
    'Gender', 'MaritalStatus_Clean', 'Monthly Income', 'MonthlyIncome_Lakhs',
    'SalaryBand', 'OverTime_Flag', 'Percent Salary Hike', 'Performance Rating',
    'PerformanceRating_Clean', 'HighPerformer', 'Total Working Years',
    'Years At Company', 'TenureGroup', 'LongTenure', 'NewHire',
    'Work Life Balance', 'WorkLifeBalance_Level', 'GoodWorkLife', 'PoorWorkLife',
    'Stock Option Level', 'HasStockOptions', 'Num Companies Worked',
    'Distance From Home', 'AvgSatisfaction', 'HighSatisfaction', 'LowSatisfaction',
    'HighInvolvement', 'LowInvolvement', 'HighPaid', 'LowPaid',
    'EmploymentStatus'
]

available = [col for col in essential if col in df_clean.columns]
df_final = df_clean[available].copy()

# Save
output = r"C:\Users\ACER\Desktop\HR_Data_Cleaned.csv"
df_final.to_csv(output, index=False)
print(f"✅ Saved to: {output}")
print(f"📈 Shape: {df_final.shape}")

# Summary
summary = pd.DataFrame({
    'Metric': ['Total Employees', 'Attrition Count', 'Attrition Rate %',
               'Avg Age', 'Avg Monthly Income', 'High Performers'],
    'Value': [total_emp, att_count, att_rate,
              df_clean['Age'].mean(), df_clean['Monthly Income'].mean(),
              df_clean['HighPerformer'].sum()]
})
summary_path = r"C:\Users\ACER\Desktop\HR_Summary.csv"
summary.to_csv(summary_path, index=False)
print(f"✅ Summary saved")


```
